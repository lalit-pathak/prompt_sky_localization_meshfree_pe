```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nLive_list = np.array([500, 1000, 1500, 2000, 2500, 3000, 3500])
start_up_times = np.array([0.85, 0.71, 0.81, 0.85, 0.78, 0.86, 0.76])
rbf_PE_times = np.array([1.45, 2.92, 4.40, 5.89, 7.31, 8.67, 10.24])

total_time = start_up_times + rbf_PE_times

fig, ax = plt.subplots(1, 1, figsize=(5, 4))

fontsize = 12
lw = 2

ax.plot(nLive_list, total_time, '-o', lw=lw, c='dodgerblue', label='total time')
ax.plot(nLive_list, start_up_times, '-o', lw=lw, c='grey', label='start-up time')
ax.plot(nLive_list, rbf_PE_times, '-o', lw=lw, c='k', label='sampling time')
ax.set_xticks([500, 1000, 1500, 2000, 2500, 3000, 3500])
ax.tick_params(labelsize=fontsize)
ax.set_xlabel(r'nLive', fontsize=fontsize)
ax.set_ylabel(r'time [mins]', fontsize=fontsize)
ax.legend(loc=2, fontsize=fontsize)
ax.grid()
fig.tight_layout()
fig.show()
```
