```python
fontsize = 14

#-- plotting arguments ----
axes_args = {}
axes_args['projection'] = 'astro hours'
axes_args['projection'] += ' ' + 'mollweide'
# axes_args['center'] = SkyCoord('13h -40d')    # This coordinate is chosen so that the broken ring in 2 Det case is visible 
ax = plt.axes(**axes_args)
ax.grid()
ax.tick_params(which='both', labelsize=fontsize)

methods = ['PyCBC', 'Meshfree']
colors = ['crimson', 'blue']

skymap = {}
metadata = {}
nside = {}
text = {}
vals = {}
cs = {}
deg2perpix = {}
probperdeg = {}

levels = [90] # credible levels
pp = np.round([levels[0]]).astype(int)
lw = 1.5

j = 0
for m in methods:
    
    skymap[m], metadata[m] =  fits.read_sky_map('{}_skymap.fits.gz'.format(m), nest=None)
    nside[m] = ah.npix_to_nside(len(skymap[m]))
    
    #--Convert sky map from probability to probability per square degree ---
    deg2perpix[m] = ah.nside_to_pixel_area(nside[m]).to_value(u.deg**2)
    probperdeg[m] = skymap[m] / deg2perpix[m]
    
    vals[m] = 100 * postprocess.find_greedy_credible_levels(skymap[m])
    cs[m] = ax.contour_hpx((vals[m], 'ICRS'), nested=metadata[m]['nest'], colors=colors[j], linewidths=1, levels=levels)

    #-- for printing the '' % credible region on the globe ---
    fmt = '%g%%'            
    text[m] = []  
    
    #-- for area of 90% credible region ---
    ii = np.round(np.searchsorted(np.sort(vals[m]), [levels[0]]) * deg2perpix[m]).astype(int)

    for i, p in zip(ii, pp):

        text[m].append('{} {:d}% area: {:,d} deg'.format(m, p, i))    # for typing 90% credible area value in plot
        
    j += 1

#-- putting legends on the skymaps ---
ax.text(1.059, 1.40, '\n'.join(text['PyCBC']), transform=ax.transAxes, ha='right', color=colors[0], fontsize=fontsize)
ax.text(1.059, 1.29, '\n'.join(text['Meshfree']), transform=ax.transAxes, ha='right', color=colors[1], fontsize=fontsize)

#-- for Host Galaxy location ---
ax.plot_coord(SkyCoord(3.44616, -0.408084, unit='rad'), '*', markerfacecolor='gainsboro', markeredgecolor='black', markersize=2)

#-- Axes Inset ---
center = SkyCoord.from_name("NGC 4993")

ax_inset = plt.axes([1.1, 0.5, 0.5, 0.5], projection='astro zoom', center=SkyCoord(196, -19.9, unit="deg"), radius=5*u.deg)
# ax_inset.tick_params(which='both', labelsize=fontsize)
ax_inset.set_xlabel(r'$\alpha$', fontsize=fontsize)
ax_inset.set_ylabel(r'$\delta$', fontsize=fontsize)
ax_inset.tick_params(labelsize=fontsize-2)

ax.mark_inset_axes(ax_inset)
ax.connect_inset_axes(ax_inset, 'upper left')
ax.connect_inset_axes(ax_inset, 'lower left')

j = 0
for m in methods:
    ax_inset.contour_hpx((vals[m], 'ICRS'), nested=metadata[m]['nest'], colors=colors[j], linewidths=lw, levels=levels)
    j += 1

ax_inset.plot_coord(SkyCoord(3.44616, -0.408084, unit='rad'), '*', markerfacecolor='gainsboro', markeredgecolor='black', markersize=10)   # for host galaxy

plt.savefig("../figures/fig_GW170817_skymaps_PyCBC_Meshfree.png", dpi=200, bbox_inches='tight')
plt.show()
```

### Nodes distribution

```python
m1_sol, m2_sol, chi1z_sol, chi2z_sol = (np.loadtxt('../points.txt', unpack=True).T)[-1, 0:4]
mchirp_sol, mass_ratio_sol = mchirp_from_mass1_mass2(m1_sol, m2_sol), q_from_mass1_mass2(m1_sol, m2_sol)

frac = [0.0, 0.2, 0.6, 0.9]
fLow = 20

nodes = {}
gauss_nodes = {}

for i in range(len(frac)):
    
    if(frac[i] != 0.0):

        with open('pickles_files_for_nodes_dist/rbfInterpolants_gauss_{}.pickle'.format(frac[i]), 'rb') as f:
            output = pickle.load(f)
            f.close()

        temp_nodes = output['nodes']
        gauss_nodes[frac[i]] = np.array(output['gauss_nodes'])

        m1 = mass1_from_tau0_tau3(temp_nodes[:,0]/(2*np.pi*fLow), temp_nodes[:,1]/(2*np.pi*fLow), fLow)
        m2 = mass2_from_tau0_tau3(temp_nodes[:,0]/(2*np.pi*fLow), temp_nodes[:,1]/(2*np.pi*fLow), fLow)
    
        mchirp, q = mchirp_from_mass1_mass2(m1, m2), q_from_mass1_mass2(m1, m2)
        
        nodes[frac[i]] = np.column_stack((mchirp, q, temp_nodes[:,2], temp_nodes[:,3]))        
        
    else:
        
        with open('pickles_files_for_nodes_dist/rbfInterpolants_gauss_{}.pickle'.format(frac[i]), 'rb') as f:
            output = pickle.load(f)
            f.close()

            temp_nodes = output['nodes']
            
            m1 = mass1_from_tau0_tau3(temp_nodes[:,0]/(2*np.pi*fLow), temp_nodes[:,1]/(2*np.pi*fLow), fLow)
            m2 = mass2_from_tau0_tau3(temp_nodes[:,0]/(2*np.pi*fLow), temp_nodes[:,1]/(2*np.pi*fLow), fLow)

            mchirp, q = mchirp_from_mass1_mass2(m1, m2), q_from_mass1_mass2(m1, m2)

            nodes[frac[i]] = np.column_stack((mchirp, q, temp_nodes[:,2], temp_nodes[:,3])) 
            
fig, ax = plt.subplots(1, 4, figsize=(16, 4), sharey=True)

fontsize = 13

ax[0].scatter(nodes[0.0][:,0], nodes[0.0][:,1], s=10, c='gray', alpha=0.5)
ax[0].scatter(mchirp_sol, mass_ratio_sol, s=80, marker='x', c='black', lw=3)
ax[0].set_ylabel(r'$q$', fontsize=fontsize)
ax[0].set_xlabel(r'$\mathcal{M}$', fontsize=fontsize)
ax[0].set_title(r'(I) $100 \% \, $Uniform', fontsize=fontsize)
ax[0].tick_params(which='both', labelsize=fontsize)
ax[0].minorticks_on()

ax[1].scatter(nodes[0.2][:,0], nodes[0.2][:,1], s=10, c='gray', alpha=0.5)
ax[1].scatter(gauss_nodes[0.2][:,0], gauss_nodes[0.2][:,1], s=10, color='crimson')
ax[1].scatter(mchirp_sol, mass_ratio_sol, s=80, marker='x',c='black', lw=3)
ax[1].set_title(r'(II) $80 \% \, $Uniform', fontsize=fontsize)
ax[1].set_xlabel(r'$\mathcal{M}$', fontsize=fontsize)
plt.setp(ax[1].get_yticklabels(), visible=False)
ax[1].tick_params(which='both', labelsize=fontsize)
ax[1].minorticks_on()

ax[2].scatter(nodes[0.6][:,0], nodes[0.6][:,1], s=10, c='gray', alpha=0.5)
ax[2].scatter(gauss_nodes[0.6][:,0], gauss_nodes[0.6][:,1], s=10, color='crimson')
ax[2].scatter(mchirp_sol, mass_ratio_sol, s=80, marker='x',c='black', lw=3)
ax[2].set_xlabel(r'$\mathcal{M}$', fontsize=fontsize)
ax[2].set_title(r'(III) $40 \% \, $Uniform', fontsize=fontsize)
ax[2].tick_params(which='both', labelsize=fontsize)
ax[2].minorticks_on()

ax[3].scatter(nodes[0.9][:,0], nodes[0.9][:,1], s=10, c='gray', alpha=0.5)
ax[3].scatter(gauss_nodes[0.9][:,0], gauss_nodes[0.9][:,1], s=10, color='crimson')
ax[3].scatter(mchirp_sol, mass_ratio_sol, s=80, marker='x', c='black', lw=3)
ax[3].set_xlabel(r'$\mathcal{M}$', fontsize=fontsize)
ax[3].set_title(r'(IV) $10s \% \, $Uniform', fontsize=fontsize)
plt.setp(ax[3].get_yticklabels(), visible=False)
ax[3].tick_params(which='both', labelsize=fontsize)
ax[3].minorticks_on()

fig.tight_layout(pad=0.2)
plt.savefig('../figures/fig_nodes_distribution.png', dpi=200, bbox_inches='tight')
plt.show()
```

### Timing with number of CPU cores

```python
_, start, pe = np.loadtxt('timings_TaylorF2.txt', unpack=True)

start_up_time = np.zeros(8)
start_up_time_ll = np.zeros(8)
start_up_time_ul = np.zeros(8)
pe_time = np.zeros(8)
pe_time_ll = np.zeros(8)
pe_time_ul = np.zeros(8)
nProcs = np.arange(8, 65, 8)

j = 0
for i in range(8):
    
    start_up_time_ll[i], start_up_time[i], start_up_time_ul[i] = np.quantile(start[j:j+10], q=[0.05, 0.5, 0.95])
    pe_time_ll[i], pe_time[i], pe_time_ul[i] = np.quantile(pe[j:j+10], q=[0.05, 0.5, 0.95])
    
    j = j + 10

err_l = (start_up_time + pe_time) - (start_up_time_ll + pe_time_ll)
err_u = (start_up_time_ul + pe_time_ul) - (start_up_time + pe_time)

tot_time_TaylorF2 = start_up_time + pe_time

#-- reading the data file ---
asymmetric_error = np.array(list(zip(err_l,err_u))).T
nProcs = np.rint(nProcs).astype(int) 

x1 = np.linspace(7, 34); y1 = 65./x1**0.85
x2 = np.linspace(33,68); y2 = 19.7/x2**0.5

fig, ax = plt.subplots()
fontsize = 12

#ax.plot(nProcs, run_time, marker='.', markersize=5, lw=2, color='crimson', label='data points')
ax.plot(x1, y1, linewidth=0.65, color='crimson', ls='-.', label=r'$\propto 1/{N_{p}}^{0.85}$')
ax.plot(x2, y2, linewidth=0.65, color='cornflowerblue', ls='--', label=r'$\propto 1/\sqrt{N_p}$')
ax.errorbar(nProcs, start_up_time + pe_time, yerr=asymmetric_error, fmt='.', ecolor = 'gray', capsize=5, barsabove=True, label='data points')

ax.set_xscale('log', base=2)
ax.set_yscale('log', base=2)
ax.xaxis.set_ticks(nProcs)
ax.xaxis.set_ticklabels(nProcs, fontsize=fontsize)
ax.yaxis.set_ticks([2, 3, 4, 8, 6, 10, 12])
ax.yaxis.set_ticklabels([2, 3, 4, 8, 6, 10, 12], fontsize=fontsize)
ax.set_xlabel(r'Number of CPUs ($N_{p}$)', fontsize=fontsize)
ax.set_ylabel(r'Meshfree runtime (mins)', fontsize=fontsize)
ax.set_xlim(7.5,68)
ax.grid()
ax.legend(fontsize=fontsize)
fig.tight_layout()
fig.savefig('../figures/fig_timings_vs_cores.png', dpi=200, bbox_inches='tight')
fig.show()
```

### Errors and speed-up

```python
llr_keys = ['llr_rbf', 'llr_pycbc']
timing_keys = ['t_rbf', 't_pycbc']

llr_data_Tay = {}
timing_data_Tay = {}

llr_data_IMR = {}
timing_data_IMR = {}

with h5py.File('error_timing_Tay.hdf', 'r') as f:
        
    for key in llr_keys:

        llr_data_Tay[key] = f[key][()]
        
    for key in timing_keys:

        timing_data_Tay[key] = f[key][()]
        
with h5py.File('error_timing_IMR.hdf', 'r') as f:
        
    for key in llr_keys:

        llr_data_IMR[key] = f[key][()]
        
    for key in timing_keys:

        timing_data_IMR[key] = f[key][()]

speed_up_pycbc_rbf_Tay = timing_data_Tay['t_pycbc']/timing_data_Tay['t_rbf']
speed_up_pycbc_rbf_IMR = timing_data_IMR['t_pycbc']/timing_data_IMR['t_rbf']

abs_error_Tay = np.sort(abs(llr_data_Tay['llr_pycbc'] - llr_data_Tay['llr_rbf']))[0:-2]
bins = np.logspace(np.log10(abs_error_Tay.min()), np.log10(abs_error_Tay.max()), 100)

fig, ax = plt.subplots(1, 1, figsize=(5,4))
fontsize = 12

ax.hist(abs_error_Tay, bins=bins, histtype='step', lw=2, color='b')
ax.axvline(x=np.median(abs_error_Tay), lw=2, ls='--', c='crimson', \
                            label=r'median ~ $\mathcal{O}\,(10^{-2})$')
ax.set_xscale('log')
ax.xaxis.set_major_locator(mticker.LogLocator(numticks=999))
ax.xaxis.set_minor_locator(mticker.LogLocator(numticks=999, subs="auto"))
ax.tick_params(labelsize=fontsize)
ax.legend(fontsize=fontsize)
ax.set_xlabel(r'$|\Delta \,\ln \mathcal{L}|$', fontsize=fontsize)
fig.tight_layout()
fig.savefig('../figures/fig_accuracy_TaylorF2.png', dpi=200, bbox_inches='tight')
fig.show()
```

```python
def ecdf(a):
    x, counts = np.unique(a, return_counts=True)
    cusum = np.cumsum(counts)
    return x, cusum / cusum[-1]

vals_pycbc_rbf, frac_pycbc_rbf = ecdf(speed_up_pycbc_rbf_Tay)
idx = np.where(vals_pycbc_rbf > 40)[0]

fig, ax = plt.subplots(1, 1, figsize=(5, 4))

fontsize = 12

ax.set_ylabel('fraction', fontsize=fontsize)
ax.set_xlabel(r't$_{PyCBC}$$\,/\,$t$_{Meshfree}$', fontsize=fontsize)
ax.plot(vals_pycbc_rbf[idx], frac_pycbc_rbf[idx], color='crimson')
ax.tick_params(axis='x', labelsize=fontsize)
ax.tick_params(axis='y', labelsize=fontsize)
ax.axvline(x=np.median(vals_pycbc_rbf[idx]), ls='--', color='black', label='x%d'%np.median(vals_pycbc_rbf[idx]))
ax.grid()
fig.legend(loc='upper left', bbox_to_anchor=(0.15, 0.95), fontsize=fontsize, title='median speed-up', title_fontsize=fontsize)
fig.tight_layout()  # otherwise the right y-label is slightly clipped
fig.savefig('../figures/fig_speed_up_TaylorF2_meshfree.png', dpi=200, bbox_inches='tight')
fig.show()
```
