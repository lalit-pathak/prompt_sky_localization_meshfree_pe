```python
import numpy as np
import corner
import h5py
import json
import pygtc
import matplotlib.pyplot as plt

from pycbc.conversions import mass1_from_mchirp_q, mass2_from_mchirp_q
# import ligo.skymap.plot
import pycbc
from matplotlib.lines import Line2D
from scipy.stats import gaussian_kde

import matplotlib.lines as mpllines
from pycbc.conversions import mass1_from_mchirp_q, mass2_from_mchirp_q
from pycbc.waveform import waveform
from pycbc.types.frequencyseries import FrequencySeries
from pycbc.conversions import mass1_from_mchirp_eta, mass2_from_mchirp_eta
from plotting_utils import pos_samples, plot_corner_2datasets, plot_corner_3datasets, title_formats
import matplotlib.image as image
from matplotlib.offsetbox import (OffsetImage, AnnotationBbox)
```

```python
samps_rbf = pos_samples('post_samples_interp_seedPE_0_nodes_seed_9_eps_10.hdf')
samps_pycbc = pos_samples('post_samples_pycbc_seedPE_0.hdf')

def lead_tidal(lambda1, lambda2, mass1, mass2):
    
    num = (mass1 + 12*mass2)*mass1**4*lambda1 + (mass2 + 12*mass1)*mass2**4*lambda2
    denom = (mass1 + mass2)**5
    
    return (16/13)  * num/denom

levels = [0.50, 0.90]
fontsize = 16
lw = 2
mass1 = mass1_from_mchirp_q(samps_rbf[:,0], samps_rbf[:,1])
mass2 = mass2_from_mchirp_q(samps_rbf[:,0], samps_rbf[:,1])
lead_tidal_param = lead_tidal(samps_rbf[:,3], samps_rbf[:,4], mass1, mass2)

samps_rbf_eff = np.column_stack((samps_rbf[:,0], samps_rbf[:,1], samps_rbf[:,2], lead_tidal_param, samps_rbf[:,5], samps_rbf[:,6], \
                                samps_rbf[:,7], samps_rbf[:,8]))

mass1 = mass1_from_mchirp_q(samps_pycbc[:,0], samps_pycbc[:,1])
mass2 = mass2_from_mchirp_q(samps_pycbc[:,0], samps_pycbc[:,1])
lead_tidal_param = lead_tidal(samps_pycbc[:,3], samps_pycbc[:,4], mass1, mass2)

samps_pycbc_eff = np.column_stack((samps_pycbc[:,0], samps_pycbc[:,1], samps_pycbc[:,2], lead_tidal_param, samps_pycbc[:,5], samps_pycbc[:,6], \
                                samps_pycbc[:,7], samps_pycbc[:,8]))

methods = ['PyCBC', 'Meshfree waveform']
labels = [r'$\mathcal{M} \,\, (M_{\odot})$', r'$q$', r'$\chi_{eff}$', r'$\tilde{\Lambda}$', r'$\iota \,\,(rad)$', \
                              r'$\alpha \,\,(rad)$', r'$\delta \,\,(rad)$', r'$d_L \, \,(Mpc)$']
titles = [r'$\mathcal{M}$', r'$q$', r'$\chi_{eff}$', r'$\tilde{\Lambda}$', r'$\iota$', r'$\alpha$', r'$\delta$', r'$d_L$']
fmt_arr = ['.5f', '.2f', '.3f', '.2f', '.3f', '.3f', '.3f', '.1f']

samps_comb = [samps_pycbc_eff, samps_rbf_eff]

lw = 1.5
measure = 'median'
bins = 20
fontsize = 18
tlabels_pycbc = title_formats(samps_pycbc_eff, labels, titles, fmt_arr, measure=measure, bins=bins, flag='first')
tlabels_rbf = title_formats(samps_rbf_eff, labels, titles, fmt_arr, measure=measure, bins=bins, flag='second')

colors = ['crimson', 'steelblue']

tlabels = [tlabels_pycbc, tlabels_rbf]

kwargs = dict(labels=labels, titles=titles, levels=levels, bins=bins, tlabels=tlabels, labelpad=0.07, label_kwargs=dict(fontsize=fontsize-1), contour_kwargs=dict(linewidths=lw), \
                                  hist_kwargs=dict(lw=lw, density=True), title_kwargs=dict(fontsize=fontsize-1), plot_density=False,\
                                  methods=methods, plot_datapoints=False, fill_contours=True, color=colors[0])

filename = '../figures/fig_corner_GW170817_PyCBC_vs_Meshfree.png'
fig, axes = plot_corner_2datasets(samps_comb, colors=colors, linestyle=['-', '-'], \
                                                    measure=measure, filename=filename, save=True, **kwargs);
```
